**Framework to append new data daily**

```python
file_path = '/content/etf_data.xlsx'
```

```python
def fetch_new_data(ticker, start_date):
  if ticker == "NIFTY 50":
    df = yf.download("^NSEI",start = start_date,progress = False)
  else:
    df = yf.download(
        f"{ticker}.NS",
        start=start_date,
        progress=False
    )

  if df.empty:
      return df

  df.reset_index(inplace=True)
  df.columns = df.columns.get_level_values(0)

  return df
```

```python
def update_excel():
    file_exists = Path(file_path).exists()

    # Open writer in append mode
    with pd.ExcelWriter(
        file_path,
        engine="openpyxl",
        mode="a",
        if_sheet_exists="overlay"
    ) as writer:

        for ticker in tickers:
            print(f"Updating {ticker}")

            try:
                existing = pd.read_excel(file_path, sheet_name=ticker)
                last_date = pd.to_datetime(existing["Date"]).max()
                start_date = last_date + timedelta(days=1)
            except Exception:
                # Sheet doesn't exist
                existing = None
                start_date = last_date + timedelta(days=1)

            #Fetch new data
            new_data = fetch_new_data(ticker, start_date)

            if new_data.empty:
                print(f"No new data for {ticker}")
                continue

            #Combine & de-duplicate
            if existing is not None:
                combined = pd.concat([existing, new_data], ignore_index=True)
                combined.drop_duplicates(subset=["Date"], inplace=True)
            else:
                combined = new_data

            #Write back to same sheet
            combined.to_excel(
                writer,
                sheet_name=ticker,
                index=False
            )

    print("Excel update complete")
```

```python
def universe_eligibility(
    data,
    as_of_date,
    atv_threshold=5e7,
    min_years=2
):

    # Ensure datetime
    data = data.copy()
    data['Date'] = pd.to_datetime(data['Date'])

    # Use only data up to as_of_date
    data = data[data['Date'] <= as_of_date]

    # Safeguard
    if data.empty:
        return False

    # Average Trading Value
    atv = (data['Close'] * data['Volume']).mean()

    # History requirement
    first_date = data['Date'].min()
    history_ok = (as_of_date - first_date) >= timedelta(days=365 * min_years)

    return (atv >= atv_threshold) and history_ok
```

```python
def portfolio_eligibility(
    data,
    as_of_date,
    ma_window=200,
    lookback_days=126
):
    """
    Portfolio eligibility as of a given date.
    """

    data = data.copy()
    data['Date'] = pd.to_datetime(data['Date'])

    # Use only data up to as_of_date
    data = data[data['Date'] <= as_of_date]

    # Not enough history
    if len(data) < max(ma_window, lookback_days):
        return False

    # Moving average
    data['200MA'] = data['Close'].rolling(window=ma_window).mean()

    latest_close = data.iloc[-1]['Close']
    latest_200ma = data.iloc[-1]['200MA']

    ma_condition = latest_close > latest_200ma

    # 6-month return
    past_price = data.iloc[-lookback_days]['Close']
    six_month_return = (latest_close / past_price) - 1

    return_condition = six_month_return > 0

    return ma_condition and return_condition
```

```python
def market_regime_flag(
    data,
    as_of_date,
    ma_window=200
):
    data = data.copy()
    data['Date'] = pd.to_datetime(data['Date'])

    # Use only data up to the evaluation date
    data = data[data['Date'] <= as_of_date]

    # Safeguard
    if len(data) < ma_window:
        return None  # or 'UNKNOWN'

    # Moving average
    data['MA'] = data['Close'].rolling(window=ma_window).mean()

    latest_close = data.iloc[-1]['Close']
    latest_ma = data.iloc[-1]['MA']

    return 'BEAR' if latest_close < latest_ma else 'BULL'
```

```python
def get_exposure(ticker):
    """
    Returns exposure bucket for a given ETF ticker
    """

    ticker = ticker.upper()

    for exposure, tickers in EXPOSURE_MAP.items():
        if any(ticker.startswith(t) for t in tickers):
            return exposure

    return "OTHER"
```

```python
def build_exposure_dataframe(tickers):
    """
    Creates a dataframe mapping ETF -> exposure bucket
    """

    data = {}

    for t in tickers:
        data[t] = get_exposure(t)

    return data
```

```python
# Costs associated

STT_RATE = 0.001
EXCHANGE_CHARGES = 0.0000345
SEBI_CHARGES = 0.0000001
STAMP = 0.00015
GST_RATE = 0.18
```

```python
def transaction_cost(buy_turnover, sell_turnover):
    """
    Calculates total transaction cost for one portfolio rebalance
    (Indian ETFs – delivery).

    Parameters
    ----------
    buy_turnover : float
        Total value of BUY trades (₹)
    sell_turnover : float
        Total value of SELL trades (₹)

    Returns
    -------
    float
        Total transaction cost (₹)
    """

    total_turnover = buy_turnover + sell_turnover

    # ---- Charges ----
    stt = STT_RATE * total_turnover                 # STT on buy + sell
    exchange = EXCHANGE_CHARGES * total_turnover    # Exchange charges
    sebi = SEBI_CHARGES * total_turnover            # SEBI charges
    stamp = STAMP * buy_turnover                    # Stamp duty (buy-side only)

    gst = GST_RATE * (exchange + sebi)              # GST only on services

    total_cost = stt + exchange + sebi + stamp + gst

    return total_cost
```

```python
def score_etfs(etf_panel, nifty_6m_ret):
    df = etf_panel.copy()

    df['Momentum'] = (
        0.4 * df['6M_ret'].rank() +
        0.3 * df['3M_ret'].rank() +
        0.3 * df['12M_ret'].rank()
    )

    df['RS'] = (df['6M_ret'] - nifty_6m_ret).rank()
    df['Risk'] = df['3M_vol'].rank(ascending=False)

    df['FinalScore'] = (
        0.6 * df['Momentum'] +
        0.25 * df['RS'] +
        0.15 * df['Risk']
    )

    return df.sort_values('FinalScore', ascending=False)
```

```python
def get_eligible_panel(etf_data, as_of_date):
    rows = []

    for ticker, df in etf_data.items():
        if ticker == 'NIFTY 50':
            continue

        if universe_eligible(df, as_of_date) and portfolio_eligible(df, as_of_date):
            row = df[df['Date'] <= as_of_date].iloc[-1]
            rows.append((ticker, row))

    if not rows:
        return None

    panel = pd.DataFrame({k: v for k, v in rows}).T
    return panel
```

```python
def score_and_rank(panel, nifty_6m_ret):
    panel = panel.copy()

    panel['Momentum'] = (
        0.4 * panel['6M_ret'].rank(ascending=True) +
        0.3 * panel['3M_ret'].rank(ascending=True) +
        0.3 * panel['12M_ret'].rank(ascending=True)
    )

    panel['RS'] = (panel['6M_ret'] - nifty_6m_ret).rank(ascending=True)
    panel['Risk'] = panel['3M_vol'].rank(ascending=False)

    panel['FinalScore'] = (
        0.6 * panel['Momentum'] +
        0.25 * panel['RS'] +
        0.15 * panel['Risk']
    )

    return panel.sort_values('FinalScore', ascending=False)
```

```python
def get_portfolio_size(regime, risk_profile):
    """
    Returns portfolio size based on market regime and risk aversion.
    """

    BULL_LIMITS = {
        'Conservative': 5,
        'Balanced': 8,
        'Aggressive': 10
    }

    BEAR_LIMITS = {
        'Conservative': 2,
        'Balanced': 3,
        'Aggressive': 5
    }

    base_sizes = {
        "BULL": BULL_LIMITS,
        "BEAR": BEAR_LIMITS
    }

    if regime not in base_sizes:
        raise ValueError("Invalid regime")

    if risk_profile not in BULL_LIMITS:
        raise ValueError("Invalid risk profile")

    size = base_sizes[regime][risk_profile]

    return size
```

```python
def build_portfolio(ranked_df, exposure_map, regime, risk_profile):
    selected = []
    used_exposures = set()

    PORTFOLIO_SIZE = get_portfolio_size(regime,risk_profile)

    for etf in ranked_df.index:
        exp = exposure_map.get(etf, 'Other')

        # Skip if exposure already taken
        if exp in used_exposures:
            continue

        selected.append(etf)
        if exp != 'Other':
          used_exposures.add(exp)

        if len(selected) == PORTFOLIO_SIZE:
            break

    # weight = 1 / len(selected)
    # return {etf: weight for etf in selected}

    scores = ranked_df.loc[selected, 'FinalScore']

    # Ensure no negative weights (important if scores can be negative)
    scores = scores - scores.min()

    weights = scores / scores.sum()

    return weights.to_dict()
```

```python
def capped_turnover(old_port, new_port, portfolio_value):
    buy, sell = 0, 0

    diffs = {}

    for etf in set(old_port) | set(new_port):
        old_w = old_port.get(etf, 0)
        new_w = new_port.get(etf, 0)
        diff = new_w - old_w

        if abs(diff) < 0.05:
            continue

        diffs[etf] = diff

        if diff > 0:
            buy += diff * portfolio_value
        else:
            sell += abs(diff) * portfolio_value

    total_turnover = buy + sell
    cap = MAX_TURNOVER * portfolio_value

    # No cap needed
    if total_turnover <= cap:
        return diffs, buy, sell

    # Scale trades proportionally
    scale = cap / total_turnover
    buy *= scale
    sell *= scale

    diffs = {k: v * scale for k, v in diffs.items()}

    return diffs, buy, sell
```

```python
def portfolio_return(portfolio, etf_data, prev_date, curr_date):
    """
    Computes portfolio return between two rebalance dates.
    """
    ret = 0.0

    for etf, weight in portfolio.items():
        df = etf_data[etf]

        start = df[df['Date'] <= prev_date].iloc[-1]['Close']
        end = df[df['Date'] <= curr_date].iloc[-1]['Close']

        etf_ret = (end / start) - 1
        ret += weight * etf_ret

    return ret
```

```python
portfolio = {}
portfolio_value = 1_000_000
history = []
risk_profile = 'Aggressive' #can choose from Conservative, Moderate, Aggressive

prev_date = None
nifty_df = etf_data['NIFTY 50']

for dt in rebalance_dates:

    # ---- APPLY RETURNS (holding period) ----
    if prev_date is not None and portfolio:
        ret = portfolio_return(portfolio, etf_data, prev_date, dt)
        portfolio_value *= (1 + ret)

    # ---- MARKET REGIME ----
    regime = market_regime(nifty_df, dt)
    if regime is None:
        prev_date = dt
        continue

    # ---- ELIGIBLE ETFs ----
    panel = get_eligible_panel(etf_data, dt)
    if panel is None:
        prev_date = dt
        continue

    nifty_6m = nifty_df[nifty_df['Date'] <= dt].iloc[-1]['6M_ret']
    ranked = score_and_rank(panel, nifty_6m)

    # ---- TARGET PORTFOLIO ----
    new_port = build_portfolio(ranked, exposure_map, market_regime(nifty_df,dt),risk_profile)

    # ---- TURNOVER + COST ----
    diffs, buy, sell = capped_turnover(
        portfolio, new_port, portfolio_value
    )

    num_of_trades = len(diffs)
    cost = transaction_cost(buy, sell)
    portfolio_value -= cost

    # ---- UPDATE HOLDINGS ----
    portfolio = new_port
    history.append((dt, portfolio_value, num_of_trades))
    # print(ranked)
    print(portfolio,dt)

    prev_date = dt
```

```python
def compute_cagr(series):
    start = series.iloc[0]
    end = series.iloc[-1]
    years = (series.index[-1] - series.index[0]).days / 365.25
    return (end / start) ** (1 / years) - 1

def compute_volatility(series):
    returns = series.pct_change().dropna()
    return returns.std() * np.sqrt(252)

def compute_max_drawdown(series):
    cum_max = series.cummax()
    drawdown = (series - cum_max) / cum_max
    return drawdown.min()

def compute_sharpe(series):
    returns = series.pct_change().dropna()
    return (returns.mean() / returns.std()) * np.sqrt(252)
```

```python
def summarize_performance(nav_series):
    return {
        "Final Value": nav_series.iloc[-1],
        "Total Return (%)": (nav_series.iloc[-1] / nav_series.iloc[0] - 1) * 100,
        "CAGR (%)": compute_cagr(nav_series) * 100,
        "Volatility (%)": compute_volatility(nav_series) * 100,
        "Max Drawdown (%)": compute_max_drawdown(nav_series) * 100,
        "Sharpe Ratio": compute_sharpe(nav_series)
    }
```
